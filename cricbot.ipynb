prompt template

```python
from langchain.prompts import PromptTemplate
template="""you are a chatbot. These humans will ask you questions related to cricket.you have to answer the question in detail even through web scraping
Keep the answer within 4 sentences and concise.

Context: {context}
Question: {question}
Answer:

"""
prompt= PromptTemplate(template=template, input_variables=["context", "question"])
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser

rag_chain = (
  {"context": docsearch.as_retriever(),  "question": RunnablePassthrough()}
  | prompt
  | llm
  | StrOutputParser()
)
```

```python
!pip install gradio SpeechRecognition pyaudio
def speech_to_text(audio):
    # Initialize recognizer
    recognizer = sr.Recognizer()

    # Use audio file to recognize speech
    with sr.AudioFile(audio) as source:
        audio_data = recognizer.record(source)

    try:
        # Using Google's free API to convert speech to text
        text = recognizer.recognize_google(audio_data)
        return text
    except sr.UnknownValueError:
        return "Sorry, I couldn't understand the audio."
    except sr.RequestError:
        return "Sorry, the speech recognition service is down."

def chatbot(input_text):

    return f"You asked: {input_text}. \nAnswer: "+ rag_chain.invoke(input_text)
```

```python
def chatbot_interface(input_type,audio_input=None, text_input=None):

    if audio_input:
        # Convert speech to text if audio input is provided
        input_text = speech_to_text(audio_input)
    elif text_input:
        input_text = text_input
    else:
        return "Please provide either audio or text input."

    return chatbot(input_text)

# Define Gradio interface with both audio and text inputs
interface = gr.Interface(
    fn=chatbot_interface,
    inputs=[gr.Radio(["text", "voice"], label="Input Type"),
        gr.Audio( type="filepath", label="Speak to Chatbot"),  # Audio input
        gr.Textbox(label="Type your message",lines=2)  # Text input
    ],
    outputs="text",
    title="Chess ChatBot ",
    description="Ask questions about the provided context",
)

# Launch the interface
interface.launch(debug=True)
```
